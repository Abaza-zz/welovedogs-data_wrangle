```python
prdictions_df_clean_v1=prdictions_df.copy()
```

```python
actual_breed=[]
for index, row in prdictions_df_clean_v1.iterrows():
    
    if row['p1_dog'] == True:
        actual_breed.append(row['p1'])
    elif row['p2_dog'] == True:
        actual_breed.append(row['p2'])
    elif row['p3_dog'] == True:
        actual_breed.append(row['p3'])
    else:
        actual_breed.append("")
```
